# **Problem description:**
A thief enters a store and finds n precious metals in powder form.
The ith item has value vi and weight wi
. The total capacity of the knapsack is
W. The thief must decide the fraction of each item to take so that the total value
of the loot is maximized, subject to the constraint that the knapsack capacity
is not exceeded

# **Algorithm description:**
The above mentioned problem can be approached using Greedy algorithm. 
First of all the ratio of value (vi) upon weight(wi) is calculated for each item and stored in a list. The list is then sorted in the descending order. Now, the item having the highest value value/weight ratio is added first to the sack as long as the weight of the item is less than the capacity. As soon as the current capacity is less than the weight of the item, the fraction of item satisfying the capacity is added, which is the value per weight multiplied by the current capacity . At the end, the total value of the loot is returned.
